```python
class Student:     # 학생 테이블
    def __init__(self, hakbun, name, ph, email, address, regstdent=None):
        self.hakbun= hakbun           # 학번
        self.name= name               # 이름
        self.ph = ph                  # 연락처
        self.email = email            # 이메일
        self.address = address        # 주소
        self.regstdent = regstdent    # 등록된 날짜

    def setHakbun(self, hakbun):
        self.hakbun = hakbun
    def getHakbun(self):
        return self.hakbun

    def setName(self, name):
        self.name = name
    def getName(self):
        return self.name

    def setPh(self, ph):
        self.ph = ph
    def getPh(self):
        return self.ph

    def setEmail(self, email):
        self.email = email
    def getEmail(self):
        return self.email

    def setAddress(self, address):
        self.address = address
    def getAddress(self):
        return self.address

    def setRegstdent(self, regstdent):
        self.regstdent = regstdent
    def getRegstdent(self):
        return self.regstdent
```

```python
class score:     # 성적 테이블
    def __init__(self, hakbun, java, python, c_code, score_sum, score_avg, regscore=None):
        self.hakbun= hakbun             # 학번
        self.java= java                 # 자바점수
        self.python = python            # 파이썬점수
        self.c_code = c_code            # c언어점수
        self.score_sum = score_sum      # 총점
        self.score_avg = score_avg      # 평균
        self.regscore = regscore        # 등록된날자
        self.cnt = 0

    def setHakbun(self, hakbun):
        self.hakbun = hakbun
    def getHakbun(self):
        return self.hakbun

    def setJava(self, java):
        self.java = java
    def getJava(self):
        if self.java == '':
            self.java = None
        else:
            self.cnt += 1
        return self.java

    def setPython(self, python):
        self.python = python
    def getPython(self):
        if self.python == '':
            self.python = None
        else:
            self.cnt += 1
        return self.python

    def setC_code(self, c_code):
        self.c_code = c_code
    def getC_code(self):
        if self.c_code == '':
            self.c_code = None
        else:
            self.cnt += 1
        return self.c_code

    def setScore_sum(self, score_sum):
        self.score_sum = score_sum
    def getScore_sum(self):
        self.score_sum = (0 if self.java==None else int(self.java)) + (0 if self.python==None else int(self.python)) + (0 if self.c_code==None else int(self.c_code))
        return self.score_sum

    def setScore_avg(self, score_avg):
        self.score_avg = score_avg
    def getScore_avg(self):
        print(self.cnt)
        self.score_avg = self.score_sum/self.cnt
        return self.score_avg

    def setRegscore(self, regscore):
        self.regscore = regscore
    def getRegscore(self):
        return self.regscore
```

```python
class StudentDao:      # SQL 작업
    def __init__(self):
        self.db = None

    def connect(self):  # connect해줌. DB명이 exam
        self.db = MySQLdb.connect('localhost', 'root', '1234', 'exam')
    def disconnect(self):
        self.db.close() # connect를 끊음

    # student 테이블 용
    def insertSt(self, info_st):      # 학생 등록
        self.connect()
        cur = self.db.cursor()
        data = (info_st.getHakbun(), info_st.getName(), info_st.getPh(), info_st.getEmail(), info_st.getAddress())
        value_cnt = ', '.join(['%s']*len(data))   # data 갯수에 따라 출력되는 %s의 갯수가 달라진다.
        sql= 'insert into student (hakbun, name, ph, email, address) values (%s)'%(value_cnt)
        cur.execute(sql, data)       # sql실행
        # 학생 테이블에 데이터 삽입
        self.db.commit()             # 적용
        cur.close()                  # cursor 닫음
        self.disconnect()            # connect 끊음

    def selectSt_last(self):  # 학생테이블(오름차순 기준)에서 마지막 데이터에 학번 가져오는 함수
        self.connect()
        cur = self.db.cursor(MySQLdb.cursors.DictCursor)
        # select에서 데이터 뽑을 때, 모두 dic형태로 뽑을 것이다.(key-value로 하는게 가져오기 편함)
        sql = "select hakbun from student order by hakbun desc limit 1"
        cur.execute(sql)
        # 학생 테이블에 있는 모든 데이터를 학번의 내림차순. 그 중 첫번째열 hakbun 가져옴.
        row = cur.fetchone()
        # 고른 학번을 row에 할당
        cur.close()
        self.disconnect()
        return row

    def selectSt_all(self):  # 학생의 등록된 정보를 학번순 오름차순 반환
        self.connect()
        cur = self.db.cursor(MySQLdb.cursors.DictCursor)
        sql = "select * from student order by hakbun asc"
        cur.execute(sql)
        # 학생 테이블에 있는 모든 데이터를 학번의 오름차순으로 가져옴.
        row = cur.fetchall()
        # 고른 데이터를 전부 row에 할당
        cur.close()
        self.disconnect()
        return row

    def selectSt_h(self, hakbun): # 학생 테이블에서 특정 학번에 관한 데이터만 반환하는 함수
        self.connect()
        cur = self.db.cursor(MySQLdb.cursors.DictCursor)
        sql = "select * from student where hakbun=%s"
        data = (hakbun, )
        cur.execute(sql, data)
        # 학생 테이블에서 특정 학번에 관한 데이터만 가져옴.
        row = cur.fetchone()
        # 학번은 기본키이기 때문에, 특정 데이터는 1개 이하 -> fetchone 사용
        cur.close()
        self.disconnect()
        return row

    def updateSt(self, tuple_data): # 학생정보 수정
        self.connect()
        cur = self.db.cursor()
        sql = "update student set name=%s, ph=%s, email=%s, address=%s where hakbun=%s"
        cur.execute(sql, tuple_data)
        # 학생 테이블에 있는 특정 학번의 데이터를 수정
        self.db.commit()
        cur.close()
        self.disconnect()

    def deleteSt(self, hakbun):  # 학생 정보 삭제
        self.connect()
        cur = self.db.cursor()
        data = (hakbun, )
        sql= 'delete from student where hakbun=%s'
        cur.execute(sql, data)
        # 힉셍 테이블에서 특정 학번이 있는 데이터(행) 삭제
        self.db.commit()
        cur.close()
        self.disconnect()

    # score 테이블 용
    def insertSc(self, info_sc): # 학생점수 등록
        self.connect()
        cur = self.db.cursor()
        data = (info_sc.getHakbun(), info_sc.getJava(), info_sc.getPython(), info_sc.getC_code(), info_sc.getScore_sum(), info_sc.getScore_avg())
        value_cnt = ', '.join(['%s']*len(data))
        sql= 'insert into score (hakbun, java, python, c_code, score_sum, score_avg) values (%s)'%(value_cnt)
        cur.execute(sql, data)
        # 성적 테이블에 데이터 삽입
        self.db.commit()
        cur.close()
        self.disconnect()

    def selectSc_all(self):  # 성적 테이블에 모든 데이터를 반환
        self.connect()
        cur = self.db.cursor(MySQLdb.cursors.DictCursor)
        sql = "select * from score order by hakbun asc"
        cur.execute(sql)
        # 성적 테이블에 모든 데이터를 학번순(오름차순)으로 반환
        row = cur.fetchall()
        cur.close()
        self.disconnect()
        return row

    def selectSc_h(self, hakbun):  # 성적테이블에서 특정 학번에 대한 데이터 반환
        self.connect()
        cur = self.db.cursor(MySQLdb.cursors.DictCursor)
        sql = "select * from score where hakbun=%s"
        data = (hakbun, )
        cur.execute(sql, data)
        # 성적테이블에서 특정 학번에 대한 데이터 반환
        row = cur.fetchone()
        # 성적테이블에 학번은 'unique'이므로, 특정 데이터는 1개 이하 -> fetchone 사용
        cur.close()
        self.disconnect()
        return row

    def updateSc(self, info_sc):  # 학생점수 수정
        self.connect()
        cur = self.db.cursor()
        data = (info_sc.getJava(), info_sc.getPython(), info_sc.getC_code(), info_sc.getScore_sum(), info_sc.getScore_avg(), info_sc.getRegscore(), info_sc.getHakbun())
        sql = "update score set java=%s, python=%s, c_code=%s, score_sum=%s, score_avg=%s, regscore=%s where hakbun=%s"
        cur.execute(sql, data)
        # 성적 테이블에 있는 특정 학번의 데이터를 수정
        self.db.commit()
        cur.close()
        self.disconnect()

    def deleteSc(self, hakbun):  # 학생점수 삭제
        self.connect()
        cur = self.db.cursor()
        data = (hakbun, )
        sql= 'delete from score where hakbun=%s'
        cur.execute(sql, data)
        # 성적 테이블에서 특정 학번이 있는 데이터(행) 삭제
        self.db.commit()
        cur.close()
        self.disconnect()

    # studnet X score   (조인용)
    def selectJoin(self):           # 학생테이블과 성적테이블을 학번으로 조인(left join)
        self.connect()
        cur = self.db.cursor(MySQLdb.cursors.DictCursor)
        sql = "select student.hakbun, name, ph, email, address, java, python, c_code, score_sum, score_avg from student left join score on student.hakbun = score.hakbun order by hakbun asc"
        cur.execute(sql)
        # left join을 사용하여 학생테이블에서 성적테이블의 성적관련 필드들만 추가됐다.(성적 데이터 없다면, NULL값이 넣어짐.)
        row = cur.fetchall()
        cur.close()
        self.disconnect()
        return row

    def selectJoin_h(self, hakbun): # 학생테이블과 성적테이블을 특정 학번으로 조인(left join)
        self.connect()
        cur = self.db.cursor(MySQLdb.cursors.DictCursor)
        sql = "select student.hakbun, name, ph, email, address, java, python, c_code, score_sum, score_avg from student left join score on student.hakbun = score.hakbun where student.hakbun=%s order by hakbun asc"
        data = (hakbun, )
        cur.execute(sql, data)
        # left join을 사용하여 특정학번이 있는 학생 데이터에서 성적관련 데이터만 추가(성적 데이터가 없다면, NULL값이 넣어짐.)
        row = cur.fetchone()
        cur.close()
        self.disconnect()
        return row
```

```python
class StudentService:             # 서비스 작업(총 8개: 학생 데이터 등록,출력,수정,삭제,검색 및 점수 데이터 등록 수정 삭제)
    def __init__(self):
        self.dao = StudentDao()

    def st_reg(self):         # 학생 등록
        name = input('이름 입력: ')
        ph = input('전화번호 입력: ')
        email = input('이메일 입력: ')
        address = input('주소 입력 : ')
        data = self.dao.selectSt_last()     # student 테이블의 전체 데이터 갯수 반환
        if not data:                     # 학번 데이터가 없을 시, 학번 값은 '20240001'
            hakbun = '20240001'    
        else:                              # 있을 시, 학번 값은 str(저장된학번+1)
            hakbun = str(int(data['hakbun'])+1)    # 이론 상 9999명의 학번을 저장 가능하다.
    
        info_st = Student(hakbun, name, ph, email, address)
        # student클래스의 인스턴스 생성
        self.dao.insertSt(info_st)      # 학생테이블에 입력힌 정보 등록

    def st_list(self):    # 학생정보 출력
        datas = self.dao.selectJoin()      # 학생과 성적 테이블이 left join으로 연결되는 함수를 호출 
        if datas:   # 학생 테이블에 더이터가 있는지 판단(없으면, 메세지 출력하고 메뉴로)
            for data in datas:
                if data['score_avg']!=None:
                    # 성적테이블이 존재하면, 성적 값도 출력
                    print('학번: {}, 이름: {}, 전화번호: {}\n  이메일: {}, 주소: {}\n    Java점수: {}, python점수: {}, C언어점수: {}, 총합: {}, 평균: {:0.2f}'.format(data['hakbun'], data['name'], data['ph'], data['email'], data['address'], data['java'], data['python'], data['c_code'], data['score_sum'], data['score_avg']))
                else:
                    # 성적테이블이 존재하지 않으면, 성적 값 생략
                    print('학번: {}, 이름: {}, 전화번호: {}\n  이메일: {}, 주소: {}'.format(data['hakbun'], data['name'], data['ph'], data['email'], data['address']))
        else:
            print('학생 데이터가 없습니다. 등록해주시길 바랍니다.')

    def st_update(self):       # 학생정보 수정
        if self.dao.selectSt_all():  # 학생 테이블에 더이터가 있는지 판단(없으면, 메세지 출력하고 메뉴로)
            hakbun = input('수정할 학번 입력: ')
            exist = self.dao.selectSt_h(hakbun)
            # 학생 테이블에 수정할 학번이 있는지 확인
            if exist:
                name = input('이름 입력: ')
                ph = input('전화번호 입력: ')
                email = input('이메일 입력: ')
                address = input('주소 입력 : ')
                tuple_data = (name, ph, email, address, hakbun)
                # sql 작업쪽에서 튜플로 데이터를 묶을 예정이므로, 그냥 여기서 튜플로 묶었다.)
                self.dao.updateSt(tuple_data)
            else:
                print('학번이 {}인 학생이 존재하지 않습니다.'.format(hakbun))
        else:
            print('학생 데이터가 없습니다. 등록해주시길 바랍니다.')

    def st_find(self):       # 학생정보 검색
        if self.dao.selectSt_all():  # 학생 테이블에 더이터가 있는지 판단(없으면, 메세지 출력하고 메뉴로)
            hakbun = input('학번 검색: ')
            exist = self.dao.selectSt_h(hakbun)
            # 학생 테이블에 검색할 학번이 있는지 확인
            if exist:
                data = self.dao.selectJoin_h(hakbun)
                if data['score_avg']!=None:
                    # 성적테이블이 존재하면, 성적 값도 출력
                    print('학번: {}, 이름: {}, 전화번호: {}\n  이메일: {}, 주소: {}\n    Java점수: {}, python점수: {}, C언어점수: {}, 총합: {}, 평균: {:0.2f}'.format(data['hakbun'], data['name'], data['ph'], data['email'], data['address'], data['java'], data['python'], data['c_code'], data['score_sum'], data['score_avg']))
                else:
                    # 성적테이블이 존재하지 않으면, 성적 값 생략
                    print('학번: {}, 이름: {}, 전화번호: {}\n  이메일: {}, 주소: {}'.format(data['hakbun'], data['name'], data['ph'], data['email'], data['address']))
            else:
                print('학번이 {}인 학생이 존재하지 않습니다.'.format(hakbun))
        else:
            print('학생 데이터가 없습니다. 등록해주시길 바랍니다.')

    def st_delete(self):     # 학생정보 삭제
        if self.dao.selectSt_all():
            hakbun = input('삭제할 학번 입력: ')
            exist = self.dao.selectSt_h(hakbun) # 학생 테이블에서 특정학번이 있는 데이터 반환
            if exist:
                exist = self.dao.selectSc_h(hakbun) # 성적테이블에서 특정학번이 있는 데이터 반환
                # 학생테이블 데이터 삭제하기전, 성적테이블에 참조되있는 데이터가 있는지 확인)
                if exist:
                    self.dao.deleteSc(hakbun)    # 참조되있는 테이블쪽 데이터 삭제
                self.dao.deleteSt(hakbun)    # 그 후, 기본키 테이블쪽 데이터 삭제
            else:
                print('학번이 {}인 학생이 존재하지 않습니다.'.format(hakbun))
        else:
            print('학생 데이터가 없습니다. 등록해주시길 바랍니다.')

    def sc_reg(self):          # 학생점수 등록
        if self.dao.selectSt_all(): # hakbun이 학생테이블에 있는가(성적테이블에 학번은 참조키이므로 확인해야 함)
            hakbun = input('점수 등록할 학번 입력: ')
            exist = self.dao.selectSt_h(hakbun)
            if exist:
                exist = self.dao.selectSc_h(hakbun)
                if not exist:              # 성적테이블에 같은 hakbun이 등록되어 있는가
                    java = input('자바점수 입력: ')
                    python = input('파이썬점수 입력: ')
                    c_code = input('C언어점수 입력 : ')
                    score_sum = None
                    score_avg = None
                
                    info_sc = score(hakbun, java, python, c_code, score_sum, score_avg)
                    # score클래스의 인스턴스 생성
                    self.dao.insertSc(info_sc) # 성적테이블에 입력한 정보 등록
                else:
                    print('이미 학번이 {}인 학생의 점수가 등록되어 있습니다.'.format(hakbun))
            else:
                print('학번이 {}인 학생이 존재하지 않습니다.'.format(hakbun))
        else:
            print('학생 데이터가 없습니다. 등록해주시길 바랍니다.')

    def sc_update(self):       # 학생점수 수정
        if self.dao.selectSc_all(): # 성적 테이블에 더이터가 있는지 판단(없으면, 메세지 출력하고 메뉴로)
            hakbun = input('수정할 학번 입력: ')
            exist = self.dao.selectSc_h(hakbun)
            # 수정할 학번이 성적 테이블에 존재하는가
            if exist:
                java = input('자바점수 입력: ')
                python = input('파이썬점수 입력: ')
                c_code = input('C언어점수 입력 : ')
                score_sum = None
                score_avg = None
                
                info_sc = score(hakbun, java, python, c_code, score_sum, score_avg, exist['regscore'])
                self.dao.updateSc(info_sc)
            else:
                print('학번이 {}인 학생이 존재하지 않습니다.'.format(hakbun))
        else:
            print('점수 데이터가 없습니다. 등록해주시길 바랍니다.')

    def sc_delete(self):       # 학생점수 삭제
        if self.dao.selectSc_all():  # 성적 테이블에 더이터가 있는지 판단(없으면, 메세지 출력하고 메뉴로)
            hakbun = input('삭제할 학번 입력: ')
            exist = self.dao.selectSc_h(hakbun)
            if exist:                      # 성적테이블에 특정학번이 있는지 확인
                self.dao.deleteSc(hakbun)  # 있다면 삭제
            else:
                print('학번이 {}인 학생이 존재하지 않습니다.'.format(hakbun))
        else:
            print('점수 데이터가 없습니다. 등록해주시길 바랍니다.')
```

```python
class Menu:       # 메뉴 작업
    def __init__(self):
        self.service = StudentService()

    def run(self):
        while True:
            try:
                menu = int(input('1.학생등록 | 2.학생리스트 | 3.학생정보수정 | 4.학생정보검색 | 5.학생정보삭제\n6.점수등록 | 7.점수수정 | 8.점수삭제 | 0.종료\n선택 : '))
                if menu == 1:                 # 학생등록
                    self.service.st_reg()
                    print('학생정보 등록 완료\n')
                elif menu == 2:               # 학생리스트
                    self.service.st_list()
                    print()
                elif menu == 3:               # 학생정보수정
                    self.service.st_update()
                    print('학생정보 수정 완료\n')
                elif menu == 4:               # 학생정보검색
                    self.service.st_find()
                    print()
                elif menu == 5:               # 학생정보삭제
                    self.service.st_delete()
                    print('학생정보 삭제 완료\n')
                elif menu == 6:               # 점수등록
                    self.service.sc_reg()
                    print('점수 등록 완료\n')
                elif menu == 7:               # 점수수정
                    self.service.sc_update()
                    print('점수 수정 완료\n')
                elif menu == 8:               # 점수삭제
                    self.service.sc_delete()
                    print('점수 삭제 완료\n')
                elif menu == 0:               # 종료
                    print('프로그램이 종료되었습니다.')
                    break
                else:
                    print('다시 번호를 입력해주세요')
            except Exception as e:
                print(e)
                print('다시 입력하세요.\n')
```

```python
start = Menu()
start.run()
```
